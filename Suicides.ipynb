```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cases = pd.read_csv('Suicides in India 2001-2012.csv')
cases
```

```python
cases.State.unique()
```

```python
cases.Type_code.unique()
```

```python
cases.Type.unique()
```

```python
cases.Age_group.unique()
```

```python
state_cases = cases[~cases.State.isin(['Total (All India)', 'Total (States)', 'Total (Uts)'])]
state_cases
```

```python
statewise = state_cases.groupby(['State','Year'], as_index=False).sum()
statewise
```

```python
all_states = statewise.State.unique()
all_states
```

```python
all_years = np.arange(2001, 2013)
all_years
```

```python
plt.figure(figsize=(15,15))
for state in all_states:
    state_list = statewise[statewise['State']==state]
    plt.plot(state_list['Year'], state_list['Total'], label=state)
    plt.legend(loc='upper left')
plt.show()
```

We can see that Andhra pradesh, Karnataka, Westbengal has consistently high suicide cases, this can be attributed to the fact of rapid urbanization and race to survival in the economy.But unpredictably, west bengal saw lesser cases in the last year, while andhra pradesh rising in numbers.

```python
states = state_cases.groupby(['State'], as_index=False).sum()
avgcases = states.Total.mean()
avgs = [avgcases for i in range(35)]
states
```

```python
plt.figure(figsize=(15,15))
plt.bar(states.State, states.Total)
plt.plot(states.State, avgs)
plt.xticks(rotation=90)
plt.show()
```

We can see a correlation between the states economy and the number of cases
